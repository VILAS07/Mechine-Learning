```python
sentence1='''CRIPTION
    The Natural Language Toolkit (NLTK) is an open source Python library
    for Natural Language Processing.  A free online book is available.'''
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
stop=stopwords.words('english')
from nltk.tokenize import word_tokenize
tok=word_tokenize(sentence1)
remove_stop=[i for i in tok if i.lower() not in stop]
remove_stop
```

#n_gram

```python
from nltk.util import ngrams
gram=ngrams(sequence=nltk.word_tokenize(sentence),n=2)
for i in gram:
    print(i)
```

#Stemming

```python
from nltk.stem import PorterStemmer 
data=PorterStemmer()
words=['programming','programmers','reaching','walking','running','eating','reaches']
for i in words:
    print(i,':',data.stem(i))

```

```python
#SnowballStemmer
from nltk.stem import SnowballStemmer
data1=SnowballStemmer('english')
for i in words:
    print(i,':',data1.stem(i))
```

#Lemmatization

```python
#nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
data2=WordNetLemmatizer()
for i in words:
    print(i,':',data2.lemmatize(i))
```
