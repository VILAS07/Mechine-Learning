```python
label=['facecream','facewash','toothpaste','bathsoap',	'shampoo','moisturizer']
x = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['toothpaste'].sum(),
    df['bathsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()]
plt.pie(x,labels=label)
plt.show()
```

```python
x = df['month_number']

plt.figure()

# Subplot 1 – Toothpaste
y = df['toothpaste']
plt.subplot(3, 1, 1)
plt.plot(x, y, marker='o')
plt.title('Toothpaste Sales')
plt.xlabel('Month Number')
plt.ylabel('Units')
plt.grid(True)

# Subplot 2 – Bathsoap
y = df['bathsoap']
plt.subplot(3, 1, 2)
plt.plot(x, y, marker='o')
plt.title('Bathsoap Sales')
plt.xlabel('Month Number')
plt.ylabel('Units')
plt.grid(True)

# Subplot 3 – Total Profit
y = df['total_profit']
plt.subplot(3, 1, 3)
plt.plot(x, y, marker='o')
plt.title('Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.grid(True)

plt.tight_layout()
plt.show()
```
