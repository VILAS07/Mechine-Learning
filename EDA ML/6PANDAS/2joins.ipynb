```python
#join ----- combine df

# inner join --- returns matching rows from both df
# left join  --- returns all the rows from left table and also matching rows from right table.if no match return null
# right join --- returns all the rows from right table and also matching rows from left table.if no return null
# full join  --- returns all rows from both the df.if no return null
```

```python
import pandas as pd

df1_data = {
    "id": [1, 2, 3, 4, 5],
    "fname": ["Vilas", "Aman", "Neha", "Rahul", "Anu"],
    "age": [22, 24, 23, 25, 21]
}
df1_data
```

```python
df2_data = {
    "id": [3, 4, 5, 6, 7],
    "prof": ["Analyst", "ML Eng", "Dev", "Tester", "Manager"],
    "sal": [60000, 85000, 70000, 50000, 90000]
}


```

```python
df1 = pd.DataFrame(df1_data)
print(df1)
```

```python
df2 = pd.DataFrame(df2_data)
print(df2)
```

```python
# inner join

new_DF=pd.merge(df1,df2,on='id',how='inner')
new_DF
```

```python
# outer join
new_dfo=pd.merge(df1,df2,on='id',how='outer')
new_dfo
```
